```python
import pandas as pd
import numpy as np
import math
from PIL import Image
from scipy.signal import convolve2d as conv2
import matplotlib.pyplot as plt
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
def gauss(sigma):
    
    #...
    Gx =[]
    x=[]
    int_sigma=int(sigma)
    for el in range( -3*int_sigma, 3 *int_sigma +1):
        n = (1/((math.sqrt(2*np.pi))*int_sigma))* np.exp((-(el**2)/(2*int_sigma**2)))
        Gx.append(n)
        x.append(el)
    Gx = np.array(Gx)
    
    return Gx, x
```

```python
def gaussdx(sigma):
    
    #...
    Dx =[]
    x=[]
    int_sigma=int(sigma)
    for el in range( -3*int_sigma, 3 *int_sigma +1):
        n = (-1/((math.sqrt(2*np.pi))*int_sigma**3))*el*np.exp((-(el**2)/(2*int_sigma**2)))

        Dx.append(n)
        x.append(el)
    Dx = np.array(Dx)
    
    return Dx, x
```

```python
plt.rcParams["figure.figsize"] = (10,2)
img_imp = np.zeros([27,27])
img_imp[13, 13] = 1.0
plt.figure(6), plt.imshow(img_imp, cmap='gray')

sigma = 7.0
[Gx, x] = gauss(sigma)
[Dx, x] = gaussdx(sigma)

Gx = Gx.reshape(1, Gx.size)
Dx = Dx.reshape(1, Dx.size)
```

```python
plt.figure(7)
plt.subplot(2,3,1)
plt.imshow(conv2(conv2(img_imp, Gx, 'same'), Gx.T, 'same') , cmap='gray')
plt.subplot(2,3,2)
plt.imshow(conv2(conv2(img_imp, Gx, 'same'), Dx.T, 'same') , cmap='gray')
plt.subplot(2,3,3)
plt.imshow(conv2(conv2(img_imp, Dx.T, 'same'), Gx, 'same') , cmap='gray')
plt.subplot(2,3,4)
plt.imshow(conv2(conv2(img_imp, Dx, 'same'), Dx.T, 'same') , cmap='gray')
plt.subplot(2,3,5)
plt.imshow(conv2(conv2(img_imp, Dx, 'same'), Gx.T, 'same') , cmap='gray')
plt.subplot(2,3,6)
plt.imshow(conv2(conv2(img_imp, Gx.T, 'same'), Dx, 'same') , cmap='gray')
plt.show()
```
